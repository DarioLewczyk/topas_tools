```python
import sys, os
import numpy as np
import plotly.express as xp
from pymatgen.core.structure import Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
```

## Goal: To Create a Simple Tool to Plot Structure Objects

```python
working_dir = os.getcwd()
cwo_dir = '/Users/dariolewczyk/Documents/Stony_Brook/Khalifah_Research_Group/BiVO4_Work/CaWO4_Enumeration/single_crystal/output/Fixed_Output/Mixed_Occ_Output/'
os.chdir(cwo_dir)
```

```python
cwo_dict = {}
for i, f in enumerate(os.listdir()):
    if os.path.isfile:
        if f.endswith('.cif'):
            number = int(f.strip('.cif').split('_')[-1]) #get the integer number of the file. 
            struct = Structure.from_file(f)
            space_group_analyzer = SpacegroupAnalyzer(struct)
            cwo_dict[number] = {'struct':struct,'sga':space_group_analyzer}
```

```python
df = cwo_dict[22]['struct'].as_dataframe()
df
```

```python
fig = xp.scatter_3d(df, x='x', y='y', z='z', color='species')
```
